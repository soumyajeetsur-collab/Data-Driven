```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([2,5,8,9])
y = np.array([1,2,3,4])
plt.plot(x,y)
plt.grid()
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np
x = np.array([2,5,8,9])
y = np.array([1,2,3,4])
plt.plot(x,y,marker='.',ms='15',color='b',mfc='g',mec='b',linestyle='-')
plt.xlabel=("Age")
plt.ylabel=("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([2,5,8,9])
y = np.array([1,2,3,4])
plt.scatter(x,y,marker='o',color='g')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([2,5,8,7,6])
y = np.array([1,2,3,4,5])

plt.figure(figsize=(10, 7))

plt.subplot(2,3,1)
plt.plot(x,y)
plt.title('Plot 1')
plt.grid()

plt.subplot(2,3,2)
plt.plot(x,y)
plt.title('Plot 2')
plt.grid()

plt.subplot(2,3,3)
plt.plot(x,y)
plt.title('Plot 3')
plt.grid()

plt.subplot(2,3,4)
plt.plot(x,y)
plt.title('Plot 4')
plt.grid()

plt.subplot(2,3,5)
plt.plot(x,y)
plt.title('Plot 5')
plt.grid()

plt.subplot(2,3,6)
plt.plot(x,y)
plt.title('Plot 6')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4])
y=np.array([3,5,8,9])
plt.bar(x,y,color='g',width=0.4)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4])
y=np.array([3,5,8,9])
plt.barh(x,y,color='b',height=0.4)
plt.show()
```
